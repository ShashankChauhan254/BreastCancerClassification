```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
print(breast_cancer_dataset)
```

```python
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
data_frame.head()
```

```python
data_frame['label'] = breast_cancer_dataset.target
data_frame.tail()
```

```python
data_frame.describe()
```

```python
data_frame['label'].value_counts()
```

```python
data_frame.groupby('label').mean()
```

```python
X = data_frame.drop(columns = 'label', axis = 1)
y = data_frame['label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_trian_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow  import keras

tf.random.set_seed(3)
```

```python
model = keras.Sequential([keras.layers.Flatten(input_shape = (30,)), keras.layers.Dense(20, activation = 'relu'), keras.layers.Dense(40, activation = 'relu'), keras.layers.Dense(2, activation = 'sigmoid')])
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
history = model.fit(X_trian_std, y_train, validation_split = 0.1, epochs = 10)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Training Data', 'Validation Data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Training Data', 'Validation Data'], loc = 'upper right')
```

```python
loss, accuracy = model.evaluate(X_test_std, y_test)
print("Loss: ", loss)
print("Accuracy: ", accuracy)
```

```python
y_pred = model.predict(X_test_std)
print(y_pred)
```

```python
y_pred_label = [np.argmax(i) for i in y_pred]
print(y_pred_label)
```
